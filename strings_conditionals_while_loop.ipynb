## **Conditional Statements**

Python conditional statements are used to execute specific blocks of code based on whether a condition is true or false, allowing programs to make decisions and control their flow.

The main keywords are:

`if` → check a condition

`elif` → check another condition if the previous `if`/`elif` was False

`else` → run code when none of the conditions are True

**N.B.** A condition → is either `True` or `False`



**a) Basic `if` statement**

Checks one condition. If it’s `True`, it runs the code inside.

```python
age = 20
if age >= 18:
    print("You are an adult.")
```
Output:
```shell
You are an adult.
```

**b) `if-else` statement**

Runs one block of code if `True` and another block if `False`.

```python
age = 15
if age >= 18:
    print("You are an adult.")
else:
    print("You are a minor.")
```

Output:
```shell
You are a minor.
```

**c) `if-elif-else` statement (multiple conditions)**

You can use this when you have more than two choices. Python will check each condition in order and run the code for the first condition that is `True`.

```python
age = 16

if age < 13:
    print("You are a child.")
elif age < 20:
    print("You are a teenager.")
else:
    print("You are an adult.")
```

Output:
```shell
You are a teenager.
```
Another example with multiple `elif` block.

```python
age = 20

if age < 13:
    print("You are a child.")
elif age < 20:
    print("You are a teenager.")
elif age < 60:
    print("You are an adult.")
else:
    print("You are a senior.")

```

Output:
```shell
You are an adult.
```

## Nested Conditions

A nested condition means writing a conditional  (`if`, `elif` or `else`) statement inside another conditional (`if`, `elif` or `else`) block. It allows Python to check conditions step by step, one after another.

`Example 1:`

```python
age = 18
has_license = False

if age >= 18:
    if has_license:
        print("You can drive cars")
    else:
        print("You need a license to drive cars")
else:
    print("You are too young too drive!")
```
<hr style="width:50%; text-align:left; margin-left:0;">

This code clearly explains why someone cannot drive:
- Either the person is too young Or
- the person does not have a license

The above code can be converted using `and` operator

```python
age = 18
has_license = False

if age >= 18 and has_license:
    print("You need a license to drive cars")
else:
    print("You are too young or you don't have a license")
```

<hr style="width:50%; text-align:left; margin-left:0;">

This version using the `and` operator is less clear.

When the condition is false, it is not clear whether the person cannot drive because:

- they are under 18, Or

- they do not have a license.

The nested version is clearer because it explains the exact reason

`Example 2:`

```python
temperature = 8
is_sunny = True

if temperature > 15:
    print("The weather is warm enough. Let's go out.")
else:
    if is_sunny:
        print("It's cold, but sunny. Let's go out.")
    else:
        print("It's cold and not sunny, so it's risky to go outside.")
```

<hr style="width:50%; text-align:left; margin-left:0;">

This nested condition clearly explains why going outside is risky.

The above code can be converted using `or` operator.

```python
temperature = 8
is_sunny = True

if temperature > 15 or is_sunny:
    print("Let's go out.")
else:
    print("It's risky to go outside in this weather condition")
```

<hr style="width:50%; text-align:left; margin-left:0;">

This version using the `or` operator is less clear.

If the condition is false, it does not clearly explain why going outside is risky.

The nested version clearly states the reason (cold and not sunny).

**Key Takeaways**

- Use nested conditions when you want clear, step-by-step explanations

- Use `and` / `or` operators when the logic is simple and the reason does not matter

- Nested conditions are more explicit and clear, while `and` / `or` conditions are shorter but less descriptive.

## **Loop intro and `while` loop**

A loop is used when we want to repeat the same code multiple times without writing it again and again. This makes our code shorter, cleaner, and easier to manage.

A `while` loop repeats a block of code as long as a condition is `True`.

>Think of it as: “Keep doing this `while` something is `True`”

`Basic Structure:`
```python
while condition:
    # code to repeat
```

`Example:`
```python
count = 0

while count < 3:
    print("Hello World")
    count = count + 1
```
`Output:`
```shell
Hello World
Hello World
Hello World
```

```python
while condition: # loop head
    codeblock    # loop body
    codeblock    # loop body
```

````python
print("Hello World")
print("Hello World")
print("Hello World")
print("Hello World")
print("Hello World")
````

````python
counter = 0

while counter < 3:
    print("Hello world")
    counter = counter + 1
````

````python
counter = 3

while counter > 0:
    print("Hello world")
    counter = counter - 1
````

## **Infinite `while` loop and loop `break`**

#### **Infinite Loop**
```python
# condition always True
condtion = True
while condition:
    codeblock  # this code block will never stop executing since the condition is always True
```

`Example:`
```python
num = 12
while num >= 10: # this condition is always True, so the loop never breaks
    print(num)
    num = num + 2
```
Since num starts at 12 and keeps increasing, num >= 10 is always true → loop never ends.


#### **Using `break` to stop a loop**

We can use break to manually stop an infinite loop.

`Example:`

```python
num = 12

while True:  # infinite loop
    print(num)
    num = num + 2

    if num > 20:
      break       # stop the loop when condition is met
```


✅ Key Points
- `while True` creates an infinite loop
- `break` is used to exit the loop early
- Always make sure a loop has a stopping condition (or a `break`), otherwise it may run forever

````python
"""
3. 🔁 Palindrome Check

A string that reads the same forward and backward.

- racecar
- dad
- level
"""

text = "madam"

reversed_text = text[::-1]

if text == reversed_text:
  print(f"The word '{text}' is a Palindrome")
else:
  print("Not a Palindrome")
````

````python
"""
4. 🔤 Anagram Check

Two strings with same characters in different order.

- Act / Cat
- Below / Elbow
- Cinema / Iceman
- Fried / Fired
- Lemon / Melon
"""

a = "listen"
b = "silent"

if sorted(a.lower()) == sorted(b.lower()):
    print("Anagram")
else:
    print("Not Anagram")
````

````python
"""
7. 🎯 Number Guessing Game (using condition)
"""

import random

secret = random.randint(1,10)

success = False

while not success:
    guess = int(input("Plese guess a number from 1 to 10: "))
    if guess == secret:
        print("Hurray")
        success = True
    else:
        print("Try Again\n")
````

````python
"""
8. 🎯 Number Guessing Game (using break)
"""

import random

secret = random.randint(1,10)


while True:
    guess = int(input("Plese guess a number from 1 to 10: "))
    if guess == secret:
        print("Hurray")
        break   # will breakout the loop
    else:
        print("Try Again\n")
````

````python
"""
9. 🏧 ATM PIN System (with limited attempts)
"""


correct_pin = "1234"
attempts = 0
max_attempts = 3

while attempts < max_attempts:
    pin = input("Enter your ATM PIN: ")

    if pin == correct_pin:
        print("✅ Access Granted")
        break
    else:
        attempts += 1
        print(f"❌ Wrong PIN. Attempts left: {max_attempts - attempts}")
else:
    print("\n 🚫 Account temporarily blocked due to 3 failed attempts.")

````
